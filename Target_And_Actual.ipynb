# DATA PREPROCESSING

CHECKING FOR MISSING VALUE AND IT'S APPEARS TO BE ZERO MISSING VALUE

```python
data.isnull().sum()
```

```python
data.fillna(method ='ffill',inplace = True)
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

CONVERTING OBJECT DATA TYPE TO DATE TIME DATATYPE

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
```

```python
print(data['Date'])
```

EXTRACTING YEAR,MONTH,DAY AND WEEK FROM THE DATE FOR TIMESTAMP ANALYSIS

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['DayOfWeek'] = data['Date'].dt.dayofweek
```

```python
data = data.drop(columns=['Date'])
```

```python
data.head(10)
```

LABEL ENCODERS AND FIT TRANSFORM OF DATA FOR CHANGING CATERGORICAL DATA TO NUMERICAL.

```python
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
data.head(10)
```

# FEATURE ENGINEERING

```python
X = data.drop(columns=['Actual Value', 'Actual Qty'])#FEATURE VARIABLE
y_value = data['Actual Value']#TARGET VARIABLE
y_qty = data['Actual Qty']
```

```python
print(X)
```

```python
print("Actual Value and Actual Qty")
print(y_value,y_qty)
```

# DATA SPLITTING

```python
X_train, X_test, y_value_train, y_value_test, y_qty_train, y_qty_test = train_test_split(X, y_value, y_qty, test_size=0.2, random_state=42)
```

```python
print(X_train)
```

# **DATA MODELLING**

RandomForestRegressor for predicting Actual Sales
AND XGBRegressor for predicting Actual Qty

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
model_value = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model_qty = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model_value.fit(X_train, y_value_train)
model_qty.fit(X_train, y_qty_train)
```

```python
y_value_pred = model_value.predict(X_test)
y_qty_pred = model_qty.predict(X_test)
```

```python
print(y_value_pred)
```

EVALLUVATION WITH REGRESSION METRICS LIKE MAE,MSE,RMSE AND R2

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)

    print(f"Evaluation metrics for {model_name}:")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"R-squared (R2): {r2}")

evaluate_model(y_value_test, y_value_pred, 'Actual Value')
evaluate_model(y_qty_test, y_qty_pred, 'Actual Quantity')
```

THE PREDICTED MODEL ACHIEVED 76% FOR ACTUAL SALES AND 83% FOR ACTUAL QTY

```python
input_instance = {
    'Floor': 'FIRST FLOOR',
    'Division': 'Girls',
    'Section': 'Girls Inner Wear',
    'Department': 'Giw-Brief',
    'Supervisor': 'MR.DANDHAPANI',
    'Merchandiser': 'MR.MANIMARAN',
    'Target Value': 1779.4,
    'Target Qty': 31.9,
    'Year': 2024,
    'Month': 7,
    'Day': 1,
    'DayOfWeek': 3
}


#input_instance.pop('Actual Value', None)
#input_instance.pop('Actual Qty', None)

for column in ['Floor', 'Division', 'Section', 'Department', 'Supervisor', 'Merchandiser']:
    input_instance[column] = label_encoders[column].transform([input_instance[column]])[0]


input_data = pd.DataFrame(input_instance, index=[0])


predicted_value = model_value.predict(input_data)
predicted_qty = model_qty.predict(input_data)


print("Predicted Actual Value:", predicted_value[0])
print("Predicted Actual Qty:", predicted_qty[0])
```
